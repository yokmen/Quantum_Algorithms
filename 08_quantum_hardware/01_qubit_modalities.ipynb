# Qubit Modalities

## Until now, "qubit" has been an abstract two-level system. In hardware, qubits are *physical* objects: superconducting circuits, trapped ions, neutral atoms, photons, electron spins. Each platform has different gate speeds, coherence times, connectivity, and engineering challenges. This notebook surveys the main families and the trade-offs that distinguish them.

# 1. DiVincenzo's criteria

## Any physical realisation of a quantum computer should satisfy the **DiVincenzo criteria** (2000):

## 1. A scalable system with well-characterised qubits.
## 2. Ability to initialise qubits to a known fiducial state (typically $|0\rangle^{\otimes n}$).
## 3. Long enough coherence times relative to gate operation times — quantified as the **quality factor** $Q = T_2 / t_{\text{gate}}$.
## 4. A universal set of quantum gates.
## 5. Qubit-specific measurement.

## Plus two for quantum communication:

## 6. The ability to interconvert stationary and flying qubits.
## 7. The ability to faithfully transmit flying qubits between locations.

## Every modality below satisfies criteria 1–5 to varying degrees. Network applications (criteria 6–7) are still mostly research-stage and naturally favour photonic systems.

# 2. Superconducting qubits (transmons)

## **What:** lithographically patterned superconducting circuits — capacitors, inductors, and **Josephson junctions** — cooled to ~10 mK in a dilution refrigerator. The most common design is the **transmon** (Koch et al., 2007), which encodes a qubit in the two lowest energy levels of an anharmonic LC oscillator.

## **Operated by:** microwave pulses (~5 GHz) shaped on an arbitrary waveform generator and delivered through coaxial lines to the chip.

## **Numbers (2025–2026):**

## - $T_1$, $T_2$ ~ 100–500 μs (Google Willow, IBM Heron).
## - Single-qubit gate time ~ 20–40 ns; two-qubit gate ~ 50–500 ns depending on architecture.
## - Single-qubit gate fidelity > 99.9%; two-qubit ~ 99.5–99.9%.
## - Largest chips: IBM Condor (1121 qubits, 2023), Atom Computing (1180 neutral atoms, 2023). Quality > quantity now matters.

## **Strengths:** fast gates, mature fab process (uses standard semiconductor lithography), all-electrical control, biggest existing chips.

## **Weaknesses:** dilution-fridge infrastructure, fixed nearest-neighbour connectivity, sensitivity to charge/flux/photon noise, qubit-to-qubit variability.

## **Players:** IBM (Heron, Crossbill), Google (Willow, Sycamore), Rigetti, Quantum Circuits Inc., IQM, OQC, Alibaba.

# 3. Trapped ions

## **What:** individual atomic ions (Yb⁺, Ba⁺, Ca⁺) held in a Paul trap by oscillating electric fields, laser-cooled to ~ μK temperatures. The qubit lives in two long-lived internal states (hyperfine or optical clock states).

## **Operated by:** focused laser beams (or microwaves for hyperfine qubits) addressing individual ions; two-qubit gates use shared motional modes of the ion chain (Mølmer–Sørensen gate).

## **Numbers:**

## - $T_1$ ~ minutes to hours; $T_2$ ~ seconds (best in the business).
## - Single-qubit gate ~ 1–10 μs; two-qubit gate ~ 10–100 μs (slower than transmons).
## - Single-qubit fidelity > 99.99%; two-qubit fidelity > 99.9% (Quantinuum H-series).
## - All-to-all connectivity within a chain; chains of 30–60 qubits per trap zone.

## **Strengths:** the highest gate fidelities and longest coherence on the market; all-to-all connectivity; identical qubits (no fabrication variation).

## **Weaknesses:** slower gates; scaling beyond a few hundred qubits requires shuttling ions between trap zones (the **QCCD** architecture).

## **Players:** Quantinuum (Honeywell + Cambridge Quantum), IonQ, Oxford Ionics, AQT, Universal Quantum.

# 4. Neutral atoms

## **What:** neutral alkali (Rb, Cs) or alkaline-earth (Sr, Yb) atoms held in arrays of optical tweezers — focused laser spots that trap individual atoms. Qubits live in hyperfine or clock states. Two-qubit gates rely on excitation to highly excited **Rydberg states**, where atoms interact strongly via dipole-dipole coupling out to several microns.

## **Numbers:**

## - $T_1, T_2$ ~ seconds; gate times ~ μs (similar to ions).
## - Two-qubit gate fidelity ~ 99.5% and climbing (QuEra Aquila, Atom Computing).
## - **Reconfigurable connectivity**: tweezer arrays can be rearranged on-the-fly, giving effective long-range coupling.
## - Largest systems: 1180+ atoms (Atom Computing, 2023) and counting.

## **Strengths:** scales to ~1000s of qubits at room-temperature-adjacent infrastructure; reconfigurable geometry is unique among modalities; identical qubits.

## **Weaknesses:** atom loss during loading, slower gates than transmons, optical infrastructure is non-trivial.

## **Players:** QuEra, Atom Computing, Pasqal, Infleqtion (formerly ColdQuanta).

# 5. Photonic qubits

## **What:** single photons in optical modes (path, polarisation, time-bin, or frequency). Linear-optical quantum computing (KLM, 2001) uses interferometers + photon-number-resolving detectors; measurement-based / cluster-state approaches generate large entangled photonic resource states up-front.

## **Numbers:**

## - Operates at room temperature (detectors typically cryogenic).
## - Gates are deterministic only with non-linearity or probabilistic via measurement and feedforward.
## - Loss is the dominant error mode; gate fidelities depend on photon indistinguishability and detector efficiency.

## **Strengths:** natural carrier for quantum communication and networking; room-temperature operation; no decoherence from thermal environment (photons don't equilibrate).

## **Weaknesses:** deterministic two-qubit gates are hard; the resource overhead for fault tolerance is enormous; loss tolerance is the engineering challenge.

## **Players:** PsiQuantum (fusion-based fault-tolerant approach), Xanadu (continuous-variable / Gaussian boson sampling), Quandela, ORCA.

# 6. Semiconductor spin qubits

## **What:** the spin of a single electron (or nuclear spin) trapped in a quantum dot patterned in Si/SiGe, GaAs, or Si/SiO₂. Qubit lives in $|\uparrow\rangle, |\downarrow\rangle$. Operated by microwave or ESR pulses, with two-qubit interactions via exchange coupling between neighbouring dots.

## **Numbers:**

## - $T_2$ ~ ms (isotopically purified silicon).
## - Gate times ~ 10–100 ns.
## - Two-qubit fidelity ~ 99%, climbing.
## - Largest demonstrations: ~10s of dots, with industrial scaling potential via standard CMOS fabrication.

## **Strengths:** leverages mature semiconductor industry directly; potentially the most scalable in terms of qubits/area; small footprint.

## **Weaknesses:** still trailing on gate count and fidelity; charge noise from interfaces remains a challenge.

## **Players:** Intel (Tunnel Falls), Diraq, Quantum Motion, SQC, IMEC.

# 7. Topological qubits (research stage)

## **What:** an exotic qubit encoded in **non-Abelian anyons** (e.g. Majorana zero modes in superconductor-semiconductor heterostructures). Operations are *braids* — physical swaps of the anyons — that are intrinsically robust to local noise.

## **Status (2026):** Microsoft announced a "topological qubit" platform (Majorana 1, 2025) based on indium-arsenide / aluminium heterostructures. Verification of the underlying Majorana physics remains contested in the broader community.

## **Strength (if it works):** topological protection means a much higher noise threshold for fault tolerance — possibly orders of magnitude.

## **Weakness:** unproven at scale; even single-qubit demonstrations are debated.

## **Players:** Microsoft (Station Q), Nokia Bell Labs, Niels Bohr Institute.

# 8. Comparison at a glance

| Modality           | Gate time | $T_2$    | 1-Q fidelity | 2-Q fidelity | Connectivity   | Temperature |
|--------------------|-----------|----------|--------------|--------------|----------------|-------------|
| Superconducting    | 20 ns     | ~ 100 μs | 99.95%       | 99.5–99.9%   | nearest-neighb | ~ 10 mK     |
| Trapped ion        | 10 μs     | ~ s      | 99.99%+      | 99.9%+       | all-to-all     | μK (laser-cooled), room-temp lab |
| Neutral atom       | μs        | ~ s      | 99.9%        | 99.5%        | reconfigurable | μK          |
| Photonic           | ~ ns      | n/a      | high         | probabilistic | flexible      | room temp + cryo detectors |
| Semiconductor spin | 10–100 ns | ~ ms     | 99.9%        | ~ 99%        | nearest-neighb | ~ 100 mK    |
| Topological        | unknown   | "long"   | unproven     | unproven     | local          | ~ 10 mK     |

## A useful heuristic: **fidelity × gate-count budget**. With 99.9% two-qubit fidelity, a circuit can absorb ~1000 entangling gates before noise drowns out the signal. With 99.99%, that goes to ~10000. Below threshold for surface codes (~ 99% for 2-qubit), error correction kicks in and the budget becomes arbitrary at the price of qubit overhead.

# Recap

## - Five active modalities: **superconducting** (transmons), **trapped ions**, **neutral atoms**, **photonic**, **spin qubits**; plus topological as research.
## - **DiVincenzo criteria** are the checklist for any platform.
## - Trade-offs: gate speed (transmons) vs fidelity/coherence (ions, atoms) vs connectivity (ions/atoms beat transmons) vs scaling/manufacturing (silicon, transmons beat ions).
## - No clear winner yet — the race is wide open.
