# The Shor Nine-Qubit Code

## Shor's 9-qubit code (1995) was the **first quantum error-correcting code**. It corrects any single-qubit error — bit flip, phase flip, or both. It does so by **concatenating** two layers of 3-qubit codes.

# 1. The construction

## Start with the *phase-flip* code: encode one qubit into three using the basis change to $|+\rangle$:

$$ \Large  |0\rangle_L = |+++\rangle, \quad |1\rangle_L = |---\rangle. $$

## This catches single $Z$ errors (the dual of the bit-flip code).

## Then encode each of the three $|\pm\rangle$ qubits in a 3-qubit bit-flip code:

$$ \Large  |+\rangle \to \tfrac{1}{\sqrt 2}(|000\rangle + |111\rangle), \quad |-\rangle \to \tfrac{1}{\sqrt 2}(|000\rangle - |111\rangle). $$

## Total: $3 \times 3 = 9$ physical qubits per logical qubit. The two layers handle phase and bit errors independently.

# 2. Logical states

$$ \Large  |0\rangle_L = \frac{1}{2\sqrt 2}(|000\rangle + |111\rangle)(|000\rangle + |111\rangle)(|000\rangle + |111\rangle), $$
$$ \Large  |1\rangle_L = \frac{1}{2\sqrt 2}(|000\rangle - |111\rangle)(|000\rangle - |111\rangle)(|000\rangle - |111\rangle). $$

## These are highly entangled 9-qubit states.

# 3. Detecting errors

## - Within each block of 3 qubits, measure two $ZZ$ parities to detect any single $X$ error.
## - Across the three blocks, measure two $XXXXXX$ parities (each spanning 6 qubits) to detect any block-level $Z$ error.

## A $Y$ error is detected as both an $X$ and a $Z$ error on the same qubit. The total *syndrome* has 8 bits ($6 + 2$). The correction is a Pauli on a single physical qubit, determined by the syndrome.

# 4. The general lesson

## The Shor code teaches three things:

## - **Concatenation works**: combine simpler codes to correct more error types.
## - **Syndrome extraction must not reveal logical information**. The parity measurements project the noise into a Pauli error   but leave the encoded amplitude $\alpha, \beta$ untouched.
## - The **[[9, 1, 3]]** notation: 9 physical qubits, 1 logical qubit, distance 3 (corrects 1 error).

# 5. Beyond Shor

## - **[[5, 1, 3]]** (Laflamme, Miquel, Paz, Zurek): the smallest QEC code that corrects a single arbitrary error.
## - **[[7, 1, 3]] Steane code**: based on the classical Hamming(7,4) code; transversal Clifford gates.
## - **Surface codes** (notebook 5): the leading family for hardware QEC.

## All of these can be described compactly in the stabiliser formalism — coming next.

# Recap

## - 9 physical qubits per logical qubit; corrects any single-qubit error.
## - Concatenates the phase-flip code on top of three bit-flip codes.
## - Established that quantum error correction is *possible* — a paradigm-changing result of 1995.
