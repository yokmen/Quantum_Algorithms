# Surface Codes

## **Surface codes** are the leading family of quantum error-correcting codes for hardware. They have local 2D structure, high error thresholds (~1%), and are CSS — making them the practical choice for IBM, Google, AWS, and most major fault-tolerant roadmaps.

# 1. The 2D layout

## Place qubits on the vertices of a 2D lattice. Two kinds of stabilizers, alternating like a checkerboard:

## - **$X$-type plaquettes**: $X_a X_b X_c X_d$ on the 4 vertices of a face.
## - **$Z$-type plaquettes**: $Z_a Z_b Z_c Z_d$ on the 4 vertices of a face.

## Each stabilizer is a 4-body Pauli on neighbouring qubits only — measurable with one ancilla and a constant number of CNOT gates. This **locality** is what makes surface codes hardware-friendly.

# 2. Logical operators

## On a square patch with the right boundary conditions, the logical $\bar X$ is a string of $X$ operators along one boundary of the patch, and $\bar Z$ is a string of $Z$ operators along the perpendicular boundary. Distance of the code = length of the shortest such string.

## So a $d \times d$ surface-code patch encodes one logical qubit with distance $d$, using $\sim 2d^2$ physical qubits.

# 3. The threshold

## If physical error rate $p$ is below the **threshold** $p_{\text{th}} \sim 1\%$, increasing $d$ exponentially suppresses the logical error rate:

$$ \Large  p_L \;\sim\; A\, \big(p / p_{\text{th}}\big)^{d/2}. $$

## Above threshold, larger codes make things *worse*. Below threshold, larger codes make logical qubits arbitrarily reliable — this is the **threshold theorem**, the founding result of fault-tolerant quantum computing.

# 4. Resources

## Conservative resource estimates for breaking RSA-2048 with Shor on a surface-code architecture: roughly **$10^7$ physical qubits and $\sim 8$ hours of runtime**, assuming $\sim 0.1\%$ physical error rates. Most of the qubits are spent on **magic state distillation factories** to provide $T$ gates fault-tolerantly.

# 5. Current status (2026)

## - Google's *Willow* chip demonstrated a logical qubit whose error rate **decreases** as code distance increases — the first   clean *below-threshold* operation (2024).
## - IBM has roadmapped distance-7+ logical qubits via the *Heron*/*Crossbill* family.
## - Quantinuum / Atom Computing have demonstrated comparable codes on trapped ions and neutral atoms with much lower physical   error rates.

## The field is moving fast — these figures are expected to grow, not shrink.

# Recap

## - 2D layout of qubits with local 4-body stabilizers.
## - Distance-$d$ patch uses ~$2d^2$ qubits; logical error rate falls exponentially with $d$ below threshold.
## - Practical choice for hardware-targeted fault tolerance.
## - Magic state distillation is the dominant resource cost in practice.
