# Entanglement

## Two qubits are **entangled** when you cannot describe their joint state as one qubit-state times another. Entanglement is the resource behind teleportation, superdense coding, and the speedup of most quantum algorithms.

# 1. Two qubits live in $\mathbb{C}^4$

## If Alice has qubit $A$ and Bob has qubit $B$, the joint system has state space $\mathbb{C}^2 \otimes \mathbb{C}^2 = \mathbb{C}^4$. A basis is

$$ \Large  |00\rangle, |01\rangle, |10\rangle, |11\rangle, $$

## where $|ab\rangle$ is shorthand for $|a\rangle_A \otimes |b\rangle_B$. A general two-qubit state is

$$ \Large  |\psi\rangle = c_{00}|00\rangle + c_{01}|01\rangle + c_{10}|10\rangle + c_{11}|11\rangle, $$

## with $\sum |c_{ab}|^2 = 1$.

```python
import numpy as np

ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)

ket00 = np.kron(ket0, ket0)
ket01 = np.kron(ket0, ket1)
ket10 = np.kron(ket1, ket0)
ket11 = np.kron(ket1, ket1)

for name, k in [('|00>', ket00), ('|01>', ket01), ('|10>', ket10), ('|11>', ket11)]:
    print(f'{name} = {k}')
```

# 2. Product states vs entangled states

## A two-qubit state is a **product state** if it factors:

$$ \Large  |\psi\rangle = |a\rangle_A \otimes |b\rangle_B. $$

## Otherwise it is **entangled**. Example of a product state:

$$ \Large  \tfrac{1}{2}(|0\rangle + |1\rangle) \otimes (|0\rangle + |1\rangle) = \tfrac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle). $$

## Example of an entangled state (the **Bell state** $|\Phi^+\rangle$):

$$ \Large  |\Phi^+\rangle = \tfrac{1}{\sqrt{2}}(|00\rangle + |11\rangle). $$

## Try to write $|\Phi^+\rangle = (\alpha|0\rangle+\beta|1\rangle) \otimes (\gamma|0\rangle+\delta|1\rangle)$ — the coefficient of $|01\rangle$ would be $\alpha\delta$, which forces $\alpha\delta = 0$, but then either $\alpha\gamma$ or $\beta\delta$ is also zero. Contradiction.

```python
plus = (ket0 + ket1) / np.sqrt(2)
product_state = np.kron(plus, plus)
print('Product (|+>|+>):', product_state)

phi_plus = (ket00 + ket11) / np.sqrt(2)
print('Bell |Phi+>     :', phi_plus)
```

# 3. The four Bell states

## The **Bell basis** of $\mathbb{C}^4$ consists of four maximally entangled states:

$$ \Large  |\Phi^\pm\rangle = \tfrac{1}{\sqrt{2}}(|00\rangle \pm |11\rangle), \quad |\Psi^\pm\rangle = \tfrac{1}{\sqrt{2}}(|01\rangle \pm |10\rangle). $$

## We'll *prepare* them with a Hadamard + CNOT circuit in chapter 03.

# 4. Why entanglement is weird: correlated measurements

## In the Bell state $|\Phi^+\rangle = (|00\rangle + |11\rangle)/\sqrt{2}$:

## - If Alice measures her qubit and gets $0$, then Bob's qubit is *guaranteed* to be in $|0\rangle$.
## - If Alice gets $1$, Bob's qubit is in $|1\rangle$.

## Their outcomes are perfectly correlated, even if Alice and Bob are far apart. Crucially this **does not let them communicate**: Alice's outcome looks fair-coin random to her, and Bob sees fair-coin randomness too. The correlation only shows up when they compare notes (which requires classical communication). This is the content of the **no-signalling theorem**.

```python
# Simulate 10000 joint measurements of |Phi+>
probs = np.abs(phi_plus)**2   # over basis [|00>,|01>,|10>,|11>]
rng = np.random.default_rng(0)
samples = rng.choice(4, size=10_000, p=probs)
labels = ['00', '01', '10', '11']
counts = {l: 0 for l in labels}
for s in samples:
    counts[labels[s]] += 1
print(counts)
# Notice 01 and 10 never appear: perfect correlation
```

# 5. Local operations on a Bell pair

## An operator acting only on qubit $A$ becomes $A \otimes I$ on the joint space; one acting only on $B$ becomes $I \otimes B$. Local actions cannot create or destroy entanglement — only **two-qubit interactions** (like CNOT) can.

```python
X = np.array([[0,1],[1,0]], dtype=complex)
I = np.eye(2, dtype=complex)

X_on_A = np.kron(X, I)        # flip Alice's qubit
new_state = X_on_A @ phi_plus
print('(X o I)|Phi+> =', new_state)
# This is |Psi+> = (|10> + |01>)/sqrt(2) — still entangled, just relabelled
```

# 6. Why entanglement matters

## - **Resource**: teleportation and superdense coding use one Bell pair as fuel.
## - **Algorithmic power**: superposition + entanglement let amplitudes interfere across an exponentially large state space.
## - **Foundational**: Bell inequalities show that no local hidden-variable theory can reproduce these correlations —   this is the empirical content of "quantum mechanics is not classical."

# Recap

## - Joint states live in tensor products; $n$ qubits → $\mathbb{C}^{2^n}$.
## - A state is entangled iff it does **not** factor as $|a\rangle\otimes|b\rangle$.
## - Bell states are the canonical maximally-entangled two-qubit states.
## - Entangled outcomes are correlated, but cannot be used to signal.
