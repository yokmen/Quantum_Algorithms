# Linear Algebra for Quantum Computing

## Quantum mechanics is, in practice, *linear algebra over the complex numbers*. Before we touch a single qubit, we need to be fluent in vectors, matrices, inner products, and a handful of special matrix types (Hermitian, unitary).

## This notebook is a refresher with a quantum slant — everything we introduce here will reappear in the next six notebooks.

# 1. Vectors and vector spaces

## A **vector** in $\mathbb{C}^n$ is a column of $n$ complex numbers:

$$ \Large  |v\rangle = \begin{pmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{pmatrix},\quad v_i \in \mathbb{C}. $$

## We will see why we already write vectors as $|v\rangle$ (Dirac notation) in notebook 3. For now, treat $|v\rangle$ as a fancy name for a column vector.

## A **vector space** is closed under addition and scalar multiplication: if $|u\rangle, |v\rangle$ are vectors and $\alpha, \beta \in \mathbb{C}$, then $\alpha|u\rangle + \beta|v\rangle$ is also a vector.

```python
import numpy as np

u = np.array([1, 2j])           # a vector in C^2
v = np.array([3, 4])

print('u       =', u)
print('v       =', v)
print('2u + 3v =', 2*u + 3*v)
```

# 2. Matrices as linear operators

## An $m\times n$ matrix $A$ sends a vector in $\mathbb{C}^n$ to a vector in $\mathbb{C}^m$ via $|v\rangle \mapsto A|v\rangle$. It is **linear**:

$$ \Large  A(\alpha|u\rangle + \beta|v\rangle) = \alpha A|u\rangle + \beta A|v\rangle. $$

## Quantum gates are just specific square matrices acting on the state vector of a qubit system.

```python
# A famous quantum gate: the Hadamard, which takes |0> to (|0>+|1>)/sqrt(2)
H = (1/np.sqrt(2)) * np.array([[1, 1],
                                [1, -1]])
ket0 = np.array([1, 0])
print('H |0> =', H @ ket0)
```

# 3. The adjoint (conjugate transpose)

## For a matrix $A$ with complex entries, the **adjoint** $A^\dagger$ is the conjugate transpose:

$$ \Large  (A^\dagger)_{ij} = \overline{A_{ji}}. $$

## In numpy this is `A.conj().T`. The adjoint is the complex generalization of the transpose, and it is the right operation to use whenever you would have used a transpose in a real-number setting.

```python
A = np.array([[1+1j, 2],
              [0,    3j]])
print('A      =\n', A)
print('A^dag  =\n', A.conj().T)
```

# 4. Two special families of matrices

## Two matrix classes appear over and over in quantum mechanics:

## - **Hermitian**: $A^\dagger = A$. Eigenvalues are real. These represent *physical observables*   like energy or spin.
## - **Unitary**: $U^\dagger U = U U^\dagger = I$. Unitaries preserve length:   $\|U|v\rangle\| = \||v\rangle\|$. These represent *time evolution* and *quantum gates*.

## Every quantum gate you ever write down will be unitary. Every quantity you ever measure will be the eigenvalue of a Hermitian operator.

```python
# Check that the Hadamard is unitary: H^dag H should equal I
I_check = H.conj().T @ H
print('H^dag H =\n', np.round(I_check, 10))

# Check that the Pauli-X (the 'quantum NOT') is Hermitian and unitary
X = np.array([[0, 1], [1, 0]])
print('X is Hermitian?', np.allclose(X, X.conj().T))
print('X is unitary?  ', np.allclose(X.conj().T @ X, np.eye(2)))
```

# 5. Eigenvalues and eigenvectors

## $|v\rangle$ is an **eigenvector** of $A$ with **eigenvalue** $\lambda$ if

$$ \Large  A|v\rangle = \lambda |v\rangle. $$

## For Hermitian operators, eigenvalues are the possible outcomes of a measurement, and eigenvectors are the states the system collapses into after measurement (more in notebook 5).

```python
eigvals, eigvecs = np.linalg.eigh(X)   # eigh for Hermitian matrices
print('Eigenvalues of X:', eigvals)
print('Eigenvectors (columns):\n', eigvecs)
```

# 6. The tensor product

## When we combine two quantum systems, the joint state space is the **tensor product** of the parts. For column vectors $|a\rangle \in \mathbb{C}^m$ and $|b\rangle \in \mathbb{C}^n$, the tensor product $|a\rangle \otimes |b\rangle$ is the $mn$-dimensional vector

$$ \Large  |a\rangle \otimes |b\rangle = \begin{pmatrix} a_1 b_1 \\ a_1 b_2 \\ \vdots \\ a_m b_n \end{pmatrix}. $$

## Two qubits live in $\mathbb{C}^2 \otimes \mathbb{C}^2 = \mathbb{C}^4$; $n$ qubits in $\mathbb{C}^{2^n}$. That exponential is the reason quantum systems are hard to simulate classically.

```python
ket0 = np.array([1, 0])
ket1 = np.array([0, 1])

ket00 = np.kron(ket0, ket0)
ket01 = np.kron(ket0, ket1)
print('|00> =', ket00)
print('|01> =', ket01)
```

# Recap

## - Quantum states are complex column vectors; quantum gates are unitary matrices.
## - The adjoint $A^\dagger$ replaces the transpose in complex settings.
## - Hermitian = observables (real eigenvalues); Unitary = evolution (length-preserving).
## - Composite systems live in tensor products, so $n$ qubits have $2^n$ amplitudes.
