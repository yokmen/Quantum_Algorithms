# Qubits and Superposition

## A classical bit is either $0$ or $1$. A **qubit** can be any complex-linear combination of those two — that combination is what we call *superposition*, and it is the heart of every quantum speedup.

# 1. The two-dimensional state space

## A qubit's state lives in $\mathbb{C}^2$. We fix an orthonormal basis called the **computational basis**:

$$ \Large  |0\rangle = \begin{pmatrix}1\\0\end{pmatrix}, \qquad |1\rangle = \begin{pmatrix}0\\1\end{pmatrix}. $$

## A general qubit state is

$$ \Large  |\psi\rangle = \alpha |0\rangle + \beta |1\rangle, \qquad \alpha, \beta \in \mathbb{C}, $$

## subject to the **normalization condition** $|\alpha|^2 + |\beta|^2 = 1$. The numbers $\alpha$ and $\beta$ are called **amplitudes**.

```python
import numpy as np

ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)

alpha = 1/np.sqrt(3)
beta  = np.sqrt(2/3) * 1j

psi = alpha*ket0 + beta*ket1
print('|psi> =', psi)
print('|alpha|^2 + |beta|^2 =', abs(alpha)**2 + abs(beta)**2)
```

# 2. Superposition is not 'both at once'

## Saying "the qubit is both 0 and 1 at the same time" is a popular but misleading slogan. A more honest reading:

> The qubit is in a definite quantum state $\alpha|0\rangle + \beta|1\rangle$. > If we measure in the computational basis, we get outcome $0$ with probability $|\alpha|^2$ > and outcome $1$ with probability $|\beta|^2$.

## Superposition is special because *amplitudes can interfere* (cancel or reinforce). Probabilities cannot. We will use this constantly in algorithms like Grover and Shor.

# 3. The Hadamard basis

## Other orthonormal bases exist. The most useful one besides $\{|0\rangle,|1\rangle\}$ is the **Hadamard basis**:

$$ \Large  |+\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \qquad |-\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle - |1\rangle). $$

## Note that $|+\rangle$ is a superposition in the computational basis but a *basis state* in the Hadamard basis. There is no objective "the" basis — the basis is a choice of measurement.

```python
plus  = (ket0 + ket1) / np.sqrt(2)
minus = (ket0 - ket1) / np.sqrt(2)

print('|+> =', plus)
print('|-> =', minus)
print('<+|->  =', np.vdot(plus, minus))    # 0  -> orthogonal
print('<+|+>  =', np.vdot(plus, plus))     # 1  -> unit
```

# 4. Global vs relative phase

## Two state vectors that differ only by an overall complex phase $e^{i\varphi}$ are physically **indistinguishable**: every measurement on $|\psi\rangle$ and $e^{i\varphi}|\psi\rangle$ produces the same probabilities. We call $e^{i\varphi}$ a **global phase** and ignore it.

## **Relative phases**, however, are *not* ignorable. The states

$$ \Large  |+\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \qquad |-\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle - |1\rangle) $$

## differ only by the relative phase of the second amplitude, and they are *orthogonal*. Telling them apart is what the Hadamard gate is for.

```python
# Global phase: same physics
psi  = (ket0 + 1j*ket1) / np.sqrt(2)
psi2 = np.exp(1j * 0.7) * psi
print('Born probs of psi :', abs(psi)**2)
print('Born probs of psi2:', abs(psi2)**2)   # identical

# Relative phase: different physics
print('|+> probs:', abs(plus)**2)
print('|-> probs:', abs(minus)**2)
print('but <+|-> =', np.vdot(plus, minus), '(orthogonal!)')
```

# 5. Geometric picture: the Bloch sphere (preview)

## Modulo global phase, any qubit state can be written as

$$ \Large  |\psi\rangle = \cos(\theta/2)|0\rangle + e^{i\varphi}\sin(\theta/2)|1\rangle, $$

## with $\theta \in [0, \pi]$, $\varphi \in [0, 2\pi)$. These are the polar/azimuthal angles of a point on a unit sphere — the **Bloch sphere**. We will draw it properly in chapter 02.

# Recap

## - A qubit is a unit vector in $\mathbb{C}^2$.
## - Computational basis: $|0\rangle, |1\rangle$. Hadamard basis: $|+\rangle, |-\rangle$.
## - Amplitudes can interfere; probabilities cannot.
## - Global phase is unobservable; relative phase is everything.
