# Measurement and the Born Rule

## Quantum states evolve continuously and deterministically (via unitaries). Measurement is the *other* thing that happens — and it is probabilistic and irreversible. The rule that connects amplitudes to probabilities is called the **Born rule**.

# 1. The Born rule (computational basis)

## Let $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$. If we measure in the computational basis:

$$ \Large  \Pr(\text{outcome } 0) = |\alpha|^2, \qquad \Pr(\text{outcome } 1) = |\beta|^2. $$

## Normalization $|\alpha|^2 + |\beta|^2 = 1$ is precisely the statement that these probabilities sum to one. After the measurement, the state **collapses** to whichever basis vector we observed.

```python
import numpy as np

alpha, beta = 1/np.sqrt(3), np.sqrt(2/3)
psi = np.array([alpha, beta], dtype=complex)

probs = np.abs(psi)**2
print('P(0) =', probs[0])
print('P(1) =', probs[1])

# Simulate 10000 measurements
rng = np.random.default_rng(0)
samples = rng.choice([0, 1], size=10_000, p=probs)
print('Empirical:', np.bincount(samples) / 10_000)
```

# 2. The general Born rule

## Let $\{|e_i\rangle\}$ be any orthonormal basis. Measuring $|\psi\rangle$ in that basis gives outcome $i$ with probability

$$ \Large  \Pr(i) = |\langle e_i | \psi \rangle|^2, $$

## after which the state collapses to $|e_i\rangle$. (More generally, for a Hermitian observable $A$ with eigendecomposition $A = \sum_i \lambda_i |e_i\rangle\langle e_i|$, the outcome is $\lambda_i$ with that same probability.)

```python
# Measure |psi> in the Hadamard basis {|+>, |->}
ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)
plus  = (ket0 + ket1) / np.sqrt(2)
minus = (ket0 - ket1) / np.sqrt(2)

psi = (3*ket0 + 4j*ket1)
psi = psi / np.linalg.norm(psi)

amp_plus  = np.vdot(plus,  psi)
amp_minus = np.vdot(minus, psi)
print('P(+) =', abs(amp_plus)**2)
print('P(-) =', abs(amp_minus)**2)
print('sum  =', abs(amp_plus)**2 + abs(amp_minus)**2)   # 1.0
```

# 3. Projective measurement and the projector formalism

## Define projectors $P_i = |e_i\rangle\langle e_i|$. Two clean formulas:

## - **Probability**: $\Pr(i) = \langle\psi| P_i |\psi\rangle$.
## - **Post-measurement state**: $|\psi_i\rangle = \dfrac{P_i|\psi\rangle}{\sqrt{\Pr(i)}}$.

## Compare to the amplitude expressions:

$$ \Large  |\langle e_i|\psi\rangle|^2 = \langle\psi|e_i\rangle\langle e_i|\psi\rangle = \langle\psi|P_i|\psi\rangle. $$

## This is *the same Born rule*, written in operator form.

```python
P_plus  = np.outer(plus,  plus.conj())
P_minus = np.outer(minus, minus.conj())

# Born rule via projectors
print('P(+) via projector =', (psi.conj() @ P_plus  @ psi).real)
print('P(-) via projector =', (psi.conj() @ P_minus @ psi).real)

# Collapse to |+> if we observed +
collapsed = (P_plus @ psi)
collapsed /= np.linalg.norm(collapsed)
print('post-measurement state:', collapsed)
```

# 4. Why $|\alpha|^2$ and not $|\alpha|$?

## The squared modulus is the unique choice that makes probabilities behave properly under change of basis (Gleason's theorem makes this precise). It is also why the **norm-squared is preserved by unitaries** — and thus why time evolution preserves total probability:

$$ \Large  \sum_i |\langle e_i | U\psi\rangle|^2 = \langle U\psi | U\psi\rangle = \langle\psi|\psi\rangle = 1. $$

# 5. Expectation values

## Repeating a measurement many times and averaging gives the **expectation value**

$$ \Large  \langle A \rangle_\psi = \langle\psi | A | \psi\rangle = \sum_i \lambda_i\,\Pr(i). $$

## This is the bridge between quantum amplitudes and the numbers you read off a lab instrument.

```python
# Expectation of Pauli-Z in state |psi>
Z = np.array([[1, 0], [0, -1]], dtype=complex)
exp_Z = (psi.conj() @ Z @ psi).real
print('<Z>_psi =', exp_Z)

# Cross-check: probabilities * eigenvalues = +1 * P(0) + (-1) * P(1)
probs = np.abs(psi)**2
print('check  =', probs[0] - probs[1])
```

# Recap

## - **Born rule**: $\Pr(i) = |\langle e_i|\psi\rangle|^2 = \langle\psi|P_i|\psi\rangle$.
## - Measurement is **probabilistic** and **collapses** the state.
## - Unitaries preserve the norm, hence total probability.
## - Expectation values: $\langle A\rangle = \langle\psi|A|\psi\rangle$.
