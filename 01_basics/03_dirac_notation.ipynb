# Dirac (Bra-Ket) Notation

## Dirac notation is the lingua franca of quantum mechanics. Once read fluently, expressions that look like soup in matrix form become one-line statements.

# 1. Kets

## A **ket** $|\psi\rangle$ is a column vector. The label inside the ket is just a name. Common conventions:

## - $|0\rangle, |1\rangle$ — the computational basis of one qubit.
## - $|+\rangle, |-\rangle$ — the Hadamard basis.
## - $|\psi\rangle, |\phi\rangle$ — generic states.

$$ \Large  |0\rangle = \begin{pmatrix}1\\0\end{pmatrix},\quad |1\rangle = \begin{pmatrix}0\\1\end{pmatrix},\quad |\psi\rangle = \alpha|0\rangle + \beta|1\rangle = \begin{pmatrix}\alpha\\\beta\end{pmatrix}. $$

# 2. Bras

## A **bra** $\langle\psi|$ is the *adjoint* (conjugate transpose) of the corresponding ket:

$$ \Large  |\psi\rangle = \begin{pmatrix}\alpha\\\beta\end{pmatrix} \quad\Longleftrightarrow\quad \langle\psi| = \begin{pmatrix}\bar\alpha & \bar\beta\end{pmatrix}. $$

## Bras live in the **dual space**; they eat kets and return numbers.

```python
import numpy as np

def ket(*amps):
    return np.array(amps, dtype=complex)

def bra(psi):
    return psi.conj()

psi = ket(1+1j, 2)
print('|psi> =', psi)
print('<psi| =', bra(psi))
```

# 3. Inner product: $\langle\phi|\psi\rangle$

## Putting a bra next to a ket gives the inner product — a complex number:

$$ \Large  \langle\phi|\psi\rangle = \sum_i \overline{\phi_i}\, \psi_i. $$

## Notation reminders:

## - $\langle\psi|\psi\rangle = \||\psi\rangle\|^2$ is the squared norm.
## - $\langle\phi|\psi\rangle = \overline{\langle\psi|\phi\rangle}$.

```python
phi = ket(0, 1)
psi = ket(1/np.sqrt(2), 1/np.sqrt(2))

inner = bra(phi) @ psi
print('<phi|psi> =', inner)
print('|<phi|psi>|^2 =', abs(inner)**2)
```

# 4. Outer product: $|\psi\rangle\langle\phi|$

## Swap the order: a ket next to a bra is the **outer product**, which is a *matrix* (operator):

$$ \Large  |\psi\rangle\langle\phi| = \begin{pmatrix}\psi_1\\\psi_2\end{pmatrix}\begin{pmatrix}\bar\phi_1 & \bar\phi_2\end{pmatrix} = \begin{pmatrix}\psi_1\bar\phi_1 & \psi_1\bar\phi_2 \\ \psi_2\bar\phi_1 & \psi_2\bar\phi_2\end{pmatrix}. $$

## Outer products are the building blocks of operators. In particular, $|\psi\rangle\langle\psi|$ is the **projector** onto the line spanned by $|\psi\rangle$.

```python
# Outer product via numpy
ket0 = ket(1, 0)
P0 = np.outer(ket0, bra(ket0))   # projector onto |0>
print('|0><0| =\n', P0)

# Apply the projector to a state
state = ket(1/np.sqrt(2), 1/np.sqrt(2))
print('P0 |state> =', P0 @ state)
```

# 5. Completeness relation

## If $\{|e_i\rangle\}$ is an orthonormal basis, then

$$ \Large  \sum_i |e_i\rangle\langle e_i| = I. $$

## This is the resolution of the identity. Inserting it into an expression is one of the most common tricks in quantum mechanics — for example,

$$ \Large  |\psi\rangle = I |\psi\rangle = \sum_i |e_i\rangle \underbrace{\langle e_i|\psi\rangle}_{\text{amplitude}}. $$

```python
e0, e1 = ket(1, 0), ket(0, 1)
I_built = np.outer(e0, bra(e0)) + np.outer(e1, bra(e1))
print('|0><0| + |1><1| =\n', I_built.real)
```

# 6. Operators sandwiched: $\langle\phi|A|\psi\rangle$

## An expression like $\langle\phi|A|\psi\rangle$ is read right-to-left: apply $A$ to $|\psi\rangle$, then take the inner product with $|\phi\rangle$. The result is a complex number called a **matrix element** of $A$. It has two faces:

$$ \Large  \langle\phi|A|\psi\rangle = \langle\phi|(A|\psi\rangle) = (\langle\phi|A)|\psi\rangle. $$

```python
X = np.array([[0, 1], [1, 0]], dtype=complex)
print('<0|X|1> =', bra(e0) @ X @ e1)   # = 1
print('<0|X|0> =', bra(e0) @ X @ e0)   # = 0
```

# Recap

| Symbol | Object | Type |
|---|---|---|
| $\lvert\psi\rangle$ | ket | column vector |
| $\langle\psi\rvert$ | bra | row vector (conjugate of ket) |
| $\langle\phi\lvert\psi\rangle$ | inner product | complex number |
| $\lvert\psi\rangle\langle\phi\rvert$ | outer product | matrix / operator |
| $\langle\phi\rvert A\lvert\psi\rangle$ | matrix element | complex number |
