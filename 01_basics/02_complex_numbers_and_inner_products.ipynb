# Complex Numbers and Inner Products

## Quantum amplitudes are **complex numbers**, and the inner product on complex vector spaces is where all the physically meaningful quantities (probabilities, overlaps, norms) come from.

# 1. Complex numbers

## A complex number is $z = a + bi$ with $a, b \in \mathbb{R}$ and $i^2 = -1$.

## - **Conjugate**: $\bar z = a - bi$.
## - **Modulus** (absolute value): $|z| = \sqrt{a^2 + b^2} = \sqrt{z \bar z}$.
## - **Polar form**: $z = r e^{i\theta}$ where $r = |z|$ and $\theta = \arg(z)$.

## Two identities to know cold:

$$ \Large  e^{i\theta} = \cos\theta + i\sin\theta, \qquad |e^{i\theta}| = 1. $$

```python
z = 1 + 2j
print('z      =', z)
print('conj   =', z.conjugate())
print('|z|    =', abs(z))
print('|z|^2  =', (z * z.conjugate()).real)

import cmath
r, theta = cmath.polar(z)
print(f'polar  = {r:.4f} * exp(i * {theta:.4f})')
```

# 2. The inner product on $\mathbb{C}^n$

## For two column vectors $|u\rangle, |v\rangle \in \mathbb{C}^n$, the **inner product** is

$$ \Large  \langle u | v \rangle = \sum_i \overline{u_i}\, v_i. $$

## Note the conjugation on the *left* vector. It is what makes the inner product **conjugate-linear in the first slot** and linear in the second. Three properties matter:

## 1. $\langle v | v \rangle \geq 0$, with equality iff $|v\rangle = 0$.
## 2. $\langle u | v \rangle = \overline{\langle v | u \rangle}$.
## 3. $\langle u | \alpha v + \beta w \rangle = \alpha \langle u | v \rangle + \beta \langle u | w \rangle$.

```python
u = np.array([1+1j, 2])
v = np.array([0, 3j])

# Conjugate the first argument, then dot
inner = np.vdot(u, v)   # numpy's vdot does conj(u) . v
print('<u|v> =', inner)
print('<v|u> =', np.vdot(v, u))
print('conj(<v|u>) =', np.conj(np.vdot(v, u)))
```

# 3. Norm and unit vectors

## The **norm** of a vector is $\||v\rangle\| = \sqrt{\langle v | v \rangle}$. A **unit vector** has norm 1.

## Every legal quantum state is a unit vector — its amplitudes squared have to sum to 1 (we will see why in notebook 5). To turn an arbitrary nonzero vector into a state, we *normalize*:

$$ \Large  |\psi\rangle = \frac{|v\rangle}{\||v\rangle\|}. $$

```python
v = np.array([1+1j, 2, 3j])
norm = np.sqrt(np.vdot(v, v).real)
psi = v / norm
print('||v|| =', norm)
print('psi   =', psi)
print('||psi|| =', np.sqrt(np.vdot(psi, psi).real))
```

# 4. Orthogonality and orthonormal bases

## Vectors $|u\rangle$ and $|v\rangle$ are **orthogonal** if $\langle u | v \rangle = 0$. An **orthonormal basis** $\{|e_1\rangle, \dots, |e_n\rangle\}$ satisfies

$$ \Large  \langle e_i | e_j \rangle = \delta_{ij}. $$

## Any vector can be expanded as $|v\rangle = \sum_i \langle e_i | v \rangle\, |e_i\rangle$, and the coefficient $\langle e_i | v \rangle$ is exactly the *amplitude* of $|v\rangle$ in the $|e_i\rangle$ direction.

```python
# The computational basis {|0>, |1>} of one qubit
e0 = np.array([1, 0])
e1 = np.array([0, 1])

print('<e0|e0> =', np.vdot(e0, e0))
print('<e0|e1> =', np.vdot(e0, e1))
print('<e1|e1> =', np.vdot(e1, e1))
```

# 5. Cauchy–Schwarz and the meaning of overlap

## The Cauchy–Schwarz inequality says

$$ \Large  |\langle u | v \rangle| \leq \||u\rangle\| \cdot \||v\rangle\|, $$

## with equality iff $|u\rangle$ and $|v\rangle$ are scalar multiples of each other. For unit vectors, $|\langle u | v \rangle|$ lives in $[0, 1]$ and measures how "close" two states are. In quantum mechanics, $|\langle u | v \rangle|^2$ is literally the probability of obtaining outcome $u$ when measuring state $v$ in a basis containing $|u\rangle$ — the **Born rule** (notebook 5).

# Recap

## - Amplitudes are complex; the modulus squared is what becomes a probability.
## - The inner product $\langle u | v \rangle$ conjugates the left vector.
## - States are unit vectors; we always normalize.
## - $|\langle u | v \rangle|^2$ is the probability of seeing $u$ in $v$.
