# Amplitude Amplification

## Grover's algorithm generalises to **amplitude amplification**: given any quantum subroutine $A$ that produces a "good" outcome with some small probability $p$, amplitude amplification boosts the success probability to near 1 using only $O(1/\sqrt p)$ calls to $A$ — a quadratic speedup over classical repetition.

# 1. The setting

## Let $A$ be a unitary that prepares

$$ \Large  A|0\rangle = \sin\theta\,|s\rangle + \cos\theta\,|s^\perp\rangle, \qquad p = \sin^2\theta. $$

## Repeated measurement of $A|0\rangle$ would yield $|s\rangle$ after $\Theta(1/p)$ tries. The trick: avoid measuring, and instead amplify the amplitude of $|s\rangle$ before measuring.

# 2. The amplification iterate

$$ \Large  Q = -A \,(2|0\rangle\langle 0| - I)\, A^{-1}\, (2|s^\perp\rangle\langle s^\perp| - I). $$

## The first factor reflects across $|s^\perp\rangle$ — this requires a phase oracle for the "good" subspace. The second is reflection across $A|0\rangle$. Their composition rotates by $2\theta$, exactly as in Grover.

# 3. Optimal iterations

## After $k$ iterations, $\Pr(\text{good}) = \sin^2((2k+1)\theta)$. Maximised at

$$ \Large  k^\star = \Big\lfloor \frac{\pi}{4\theta} - \frac{1}{2} \Big\rfloor \approx \frac{\pi}{4\sqrt p}. $$

## If $p$ is not known precisely, one can use a randomised exponential-search strategy that still achieves $O(1/\sqrt p)$ calls.

```python
# Demonstrate amplitude amplification on a synthetic example
import numpy as np

p = 1/16                          # success probability after one call of A
theta = np.arcsin(np.sqrt(p))
iters = int(np.floor(np.pi/(4*theta) - 0.5))
print(f'p = {p}, theta = {theta:.4f}, optimal iterations = {iters}')

# After k iterations, probability becomes sin^2((2k+1) theta)
for k in range(iters + 2):
    print(f'k = {k:2d}: P(good) = {np.sin((2*k+1)*theta)**2:.4f}')
```

# 4. Applications

## Amplitude amplification appears everywhere in modern quantum algorithms:

## - **Quantum walks** on graphs (chapter 07): finding marked vertices.
## - **Quantum minimum-finding**: $O(\sqrt N)$ instead of $N$.
## - **Estimating expectation values** (with **amplitude estimation**, which extracts $\theta$ itself in $O(1/\varepsilon)$).
## - **Speeding up Monte Carlo simulation** — the basis of the quadratic speedup for many financial / pricing applications.
