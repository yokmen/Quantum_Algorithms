# Quantum Complexity Classes (BQP, QMA, ...)

## Complexity theory sorts problems by the resources needed to solve them. The quantum versions of the standard classes — **BQP** (efficient quantum), **QMA** (quantum NP), **QCMA**, **QIP** — capture exactly *what* a quantum computer buys over a classical one in the worst case. This notebook lays out the definitions, the known containments, and the conjectured separations.

# 1. The classical baseline

## A quick refresher on the classical classes that the quantum ones are defined against:

## - **P**: decidable in deterministic polynomial time on a Turing machine.
## - **BPP** (*Bounded-error Probabilistic Polynomial*): decidable in polynomial time with a coin-flipping machine,   error $\leq 1/3$ on every input. The de-facto formalisation of "efficient classical computation".
## - **NP**: decidable in polynomial time *given a polynomial-size witness*. Equivalent: a poly-time verifier accepts   on some witness for yes-instances and rejects all witnesses for no-instances.
## - **PSPACE**: decidable using a polynomial amount of memory (no time bound).
## - **PP**: probabilistic poly-time with error possibly arbitrarily close to $1/2$ — counts accepting paths.
## - **#P**: counts the number of accepting paths of an NP-machine. Sits above NP.

## The standard containment chain is

$$ \Large  \mathrm P \subseteq \mathrm{BPP} \subseteq \mathrm{NP} \subseteq \mathrm{PH} \subseteq \mathrm{PSPACE} \subseteq \mathrm{EXP}. $$

## Almost every "P vs NP"-style separation is open. Everything quantum slots into this picture.

# 2. BQP — efficient quantum computation

## **Definition.** A language $L$ is in **BQP** if there is a uniform family of polynomial-size quantum circuits $\{C_n\}$ such that for every input $x$ of length $n$:

## - $x \in L$ ⇒ $\Pr[C_n(x) \text{ accepts}] \geq 2/3$,
## - $x \notin L$ ⇒ $\Pr[C_n(x) \text{ accepts}] \leq 1/3$.

## "Uniform" means a classical poly-time Turing machine outputs the description of $C_n$ on input $1^n$. Acceptance is read off a designated output qubit.

## A few facts worth knowing:

## - **Gate set is irrelevant.** Any reasonable universal gate set (e.g. Clifford + T) yields the same class, because Solovay–Kitaev   gives a polylog overhead in $1/\varepsilon$.
## - **Error amplification.** Just like BPP, the constants $2/3$ and $1/3$ are arbitrary — any gap $\geq 1/\mathrm{poly}(n)$ can be   amplified to exponentially close to $0$ and $1$ by repeating the circuit $O(\log(1/\delta))$ times and taking the majority vote.
## - **Closed under poly-time reductions and intersection.**

# 3. What lives in BQP?

## - **FACTORING** and **DISCRETE LOG** (Shor, 1994) — believed to be outside BPP.
## - **HIDDEN SUBGROUP** for abelian groups (generalisation of Simon, Shor).
## - **PELL'S EQUATION**, principal ideal problem in number fields (Hallgren).
## - **Approximating Jones polynomials** at roots of unity (Aharonov–Jones–Landau).
## - **Simulating local Hamiltonians** for poly-time dynamics (Lloyd, Berry et al.).
## - **Solving sparse, well-conditioned linear systems** in the HHL sense.

## All of P and BPP are in BQP: a quantum circuit can simulate a classical one with a constant-factor overhead (replace AND/OR/NOT by reversible Toffoli + ancilla and discard the garbage at the end). So $\mathrm{BPP} \subseteq \mathrm{BQP}$.

# 4. Where BQP sits

## The cleanest known containment is

$$ \Large  \mathrm{BPP} \subseteq \mathrm{BQP} \subseteq \mathrm{PP} \subseteq \mathrm{PSPACE}. $$

## - $\mathrm{BQP} \subseteq \mathrm{PP}$: a classical PP-machine can sum amplitudes path-by-path, so it can compute the   acceptance probability of any quantum circuit (Adleman–DeMarrais–Huang).
## - $\mathrm{PP} \subseteq \mathrm{PSPACE}$ is classical.
## - BQP is **not** known to be in NP, and NP is **not** known to be in BQP. Both directions are wide-open.
## - **BQP vs PH** (the polynomial hierarchy) is open. Raz–Tal (2019) showed an *oracle* separation: there is an oracle relative   to which $\mathrm{BQP} \not\subseteq \mathrm{PH}$. The Forrelation problem is the witness.

## So in pictures:

```
                  PSPACE
                    |
                   PP
                  /  \
                BQP   PH
                 |     |
                BPP   NP
                  \   /
                    P
```

## with all the inclusions one-directional and (almost all) provable separations open.

# 5. QMA — the quantum analogue of NP

## **Definition.** $L \in \mathrm{QMA}$ iff there is a poly-time quantum verifier $V$ such that for every $x$:

## - $x \in L$ ⇒ $\exists$ a poly-size quantum witness $|\psi\rangle$ with $\Pr[V(x, |\psi\rangle) \text{ accepts}] \geq 2/3$,
## - $x \notin L$ ⇒ $\forall$ quantum states $|\psi\rangle$, $\Pr[V(x, |\psi\rangle) \text{ accepts}] \leq 1/3$.

## The witness is a *quantum* state (not a classical string). Otherwise the definition is exactly $\mathrm{NP}$ with a quantum verifier.

## $\mathrm{QCMA}$ is the same definition but with a **classical** witness — a poly-size bit string verified by a quantum circuit. $\mathrm{NP} \subseteq \mathrm{QCMA} \subseteq \mathrm{QMA}$.

## **QMA-complete problem.** $k$**-LOCAL HAMILTONIAN**: given a Hamiltonian $H = \sum_j H_j$ where each $H_j$ acts on at most $k$ qubits, and numbers $a < b$ with $b - a \geq 1/\mathrm{poly}(n)$, decide whether the ground-state energy is $\leq a$ or $\geq b$. This is QMA-complete for $k \geq 2$ (Kitaev's quantum Cook–Levin theorem, with sharpenings by Kempe–Kitaev–Regev).

## So $k$-LOCAL HAMILTONIAN plays the same role for QMA that 3-SAT plays for NP. It is also the *physical* statement that estimating ground-state energies of local Hamiltonians is, in the worst case, as hard as anything in QMA.

# 6. Other QMA-complete problems

## QMA-completeness is a fast-growing list. A few representative examples:

## - **LOCAL HAMILTONIAN** on various restricted lattices (2D, 1D with qudits of dimension 12 — Aharonov–Gottesman–Irani–Kempe).
## - **CONSISTENCY OF DENSITY MATRICES**: given local reduced density matrices, do they come from a global state? (Liu)
## - **QUANTUM CIRCUIT SATISFIABILITY**: does there exist an input $|\psi\rangle$ that makes a given circuit accept with high probability?
## - **APPROXIMATE GROUND-STATE ENERGY ESTIMATION** for almost any physically realistic Hamiltonian family.

## Roughly: any natural quantum optimisation problem over states ends up QMA-complete, just as natural classical CSP-style problems end up NP-complete.

# 7. Interactive proofs: QIP and MIP*

## **QIP** is the quantum analogue of $\mathrm{IP}$ — a quantum verifier exchanging quantum messages with an unbounded prover. The headline result is

$$ \Large  \mathrm{QIP} = \mathrm{PSPACE} \qquad \text{(Jain–Ji–Upadhyay–Watrous, 2010).} $$

## So quantum interaction does *not* enlarge the class beyond what classical interaction already gave (since $\mathrm{IP} = \mathrm{PSPACE}$ as well, by LFKN–Shamir). The compelling quantum feature is that $\mathrm{QIP}$ can be done in **three messages** (Kitaev–Watrous), while classical $\mathrm{IP}$ in three messages is conjecturally weaker.

## The truly dramatic class is $\mathrm{MIP}^*$: multiple provers sharing **entanglement**, no communication between provers, classical verifier. The bombshell result

$$ \Large  \mathrm{MIP}^* = \mathrm{RE} \qquad \text{(Ji–Natarajan–Vidick–Wright–Yuen, 2020).} $$

## says these interactive proofs can verify membership in *every* recursively enumerable language — including undecidable ones like the halting problem. As a corollary, the long-standing **Connes' embedding conjecture** in operator algebras is false.

# 8. Sub-universal models and sampling

## Not every quantum complexity class corresponds to "full" quantum computation. Three important sub-universal models:

## - **IQP** (*Instantaneous Quantum Polynomial-time*): circuits of the form $H^{\otimes n} \cdot D \cdot H^{\otimes n}$ where $D$ is diagonal.   Conjecturally hard to sample classically (Bremner–Jozsa–Shepherd).
## - **Boson sampling**: sampling from the output distribution of a linear-optical interferometer fed with single photons.   Aaronson–Arkhipov showed classical hardness assuming standard conjectures.
## - **Random circuit sampling**: sampling from the output of a random Clifford+T or random-2-qubit-gate circuit. The basis for   Google's 2019 "quantum supremacy" experiment.

## All three give **sampling**-based evidence for quantum advantage that does not require BQP $\neq$ BPP — they exploit the average-case hardness of computing output probabilities, which lives much higher up the hierarchy (P^#P).

````python
# A toy "verifier" for a QMA-style problem: given a 2-local Hamiltonian H = sum_j H_j,
# the YES-witness is the ground state |psi>, and acceptance reports <psi|H|psi> via Hamiltonian averaging.
import numpy as np

def random_2local_hamiltonian(n, m, rng):
    """Sum of m random Hermitian 2-local terms on n qubits, normalised."""
    H = np.zeros((2**n, 2**n), dtype=complex)
    for _ in range(m):
        i, j = sorted(rng.choice(n, size=2, replace=False))
        # random 4x4 Hermitian
        A = rng.normal(size=(4, 4)) + 1j*rng.normal(size=(4, 4))
        A = (A + A.conj().T) / 2
        # embed into n qubits
        full = np.array([[1.0]])
        for q in range(n):
            if q == i or q == j:
                continue
            full = np.kron(full, np.eye(2))
        # cheap embedding: lift A to the (i,j) qubits by a permutation
        # (acceptable for an illustrative demo; not a tight encoding)
        H += np.kron(np.eye(2**(n-2)), A) / m
    return (H + H.conj().T) / 2

rng = np.random.default_rng(42)
n, m = 4, 6
H = random_2local_hamiltonian(n, m, rng)
eigs = np.linalg.eigvalsh(H)
print(f'ground energy a(H) = {eigs[0]:.4f}')
print(f'first excited     = {eigs[1]:.4f}')
print(f'spectral gap      = {eigs[1] - eigs[0]:.4f}')
print('The QMA witness for the YES instance "a(H) <= a" is the ground state.')
````

# 9. What is known, what is not

## **Known unconditionally:**

## - $\mathrm{BPP} \subseteq \mathrm{BQP} \subseteq \mathrm{PP} \subseteq \mathrm{PSPACE}$.
## - $\mathrm{NP} \subseteq \mathrm{QCMA} \subseteq \mathrm{QMA} \subseteq \mathrm{PP}$.
## - $\mathrm{QIP} = \mathrm{IP} = \mathrm{PSPACE}$.
## - $\mathrm{MIP}^* = \mathrm{RE}$.

## **Known relative to oracles** (no unconditional proof):

## - $\mathrm{BQP} \not\subseteq \mathrm{PH}$ (Raz–Tal).
## - $\mathrm{NP} \not\subseteq \mathrm{BQP}$ (BBBV — the $\sqrt N$ Grover lower bound).

## **Wide open:**

## - $\mathrm{BPP} \stackrel{?}{=} \mathrm{BQP}$. The honest belief is "no," but a proof would imply $\mathrm{P} \neq \mathrm{PSPACE}$,   so it is not expected any time soon.
## - $\mathrm{QCMA} \stackrel{?}{=} \mathrm{QMA}$. Believed unequal; an oracle separation is known.
## - Is there a *natural* problem in BQP $\setminus$ BPP that does not look like factoring/order-finding? The hidden-subgroup problem   for non-abelian groups is the main candidate.

# Recap

## - **BQP** = bounded-error quantum poly-time. Contains BPP, factoring, discrete log; in PP and PSPACE.
## - **QMA** = quantum NP, with QMA-complete problem $k$-LOCAL HAMILTONIAN.
## - **QCMA** = quantum verifier, *classical* witness — sandwiched between NP and QMA.
## - **QIP = PSPACE**; **MIP\* = RE** (and Connes' conjecture is false).
## - Most quantum-vs-classical separations are known only relative to an oracle.
