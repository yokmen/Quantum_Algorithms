# The Quantum Circuit Model

## A **quantum circuit** is the standard recipe for writing down a quantum algorithm: a list of qubits and a list of gates applied to them in sequence, ending with measurements. Today's quantum hardware is essentially built to execute circuits — so this is the model in which most algorithms are described.

# 1. The pieces

## - **Wires** — horizontal lines, one per qubit. Time flows left to right.
## - **Gates** — boxes / symbols on the wires; each is a unitary acting on the qubits it touches.
## - **Measurements** — usually drawn at the right, as a meter symbol.
## - **Classical wires** (sometimes a double line) — carry the classical outcomes of measurements, which can control later gates (e.g. in teleportation, chapter 3 notebook 3).

# 2. Reading the circuit as a matrix product

## If a circuit applies $U_1$ then $U_2$ then $U_3$ to the initial state $|\psi_0\rangle$, the final state is

$$ \Large  |\psi_{\text{out}}\rangle = U_3\, U_2\, U_1\, |\psi_0\rangle. $$

## **Right-to-left** for the math, even though circuits are drawn **left-to-right** in time. Conversion between the two conventions trips everyone up the first time.

# 3. The standard input/output convention

## Circuits almost always start with all qubits in $|0\rangle$. So if the circuit applies unitary $U$ and then measures in the computational basis, the probability of outcome $x$ is

$$ \Large  \Pr(x) = |\langle x | U | 0\dots 0\rangle|^2. $$

## All of complexity theory and algorithm design happens inside the choice of $U$.

# 4. Tensor-product structure

## When a gate acts only on a subset of qubits, it is tensored with identity on the others. For two qubits, applying $G$ to qubit 0 alone is the operator $G \otimes I$. Applying $G$ to qubit 1 is $I \otimes G$.

## Note: numpy's `np.kron(A, B)` places $A$ on the **most significant** (leftmost in the ket) qubit. Different frameworks use different conventions — Qiskit, for example, uses *little-endian* ordering, so the wire labelled $q_0$ in a Qiskit diagram is the **rightmost** qubit in the state vector.

# 5. Depth, width, and gate count

## Three numbers we will care about for every circuit:

## - **Width** = number of qubits.
## - **Gate count** = total number of gates.
## - **Depth** = number of timesteps, where gates on disjoint qubits can run in parallel.

## Depth determines how long the circuit takes to run. On noisy hardware, depth determines how much decoherence accumulates — so reducing it is the central concern of circuit optimisation.
