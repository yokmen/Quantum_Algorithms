# 2. Your first circuit

## A `QuantumCircuit(n, m)` has $n$ qubits and $m$ classical bits. Methods like `.h(0)`, `.cx(0, 1)`, `.measure(qubit, cbit)` add gates and measurements.

````python
from qiskit import QuantumCircuit

# Bell-pair preparation circuit, plus measurements
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

print(qc)
````

# 3. Drawing

## `qc.draw('mpl')` renders the circuit as a Matplotlib figure. `qc.draw('text')` gives ASCII art (the default in print()).

# 4. Running on the Aer simulator

## The Aer simulator runs the circuit many times ("shots") and returns a histogram of measurement outcomes — exactly what real hardware would give you.

````python
from qiskit_aer import AerSimulator
from qiskit import transpile

sim = AerSimulator()
tqc = transpile(qc, sim)
result = sim.run(tqc, shots=1024).result()
counts = result.get_counts()
print('Counts:', counts)
````

# 5. Statevector inspection

## Sometimes you want the full state vector (no measurement). Build a circuit *without* measurements and use `Statevector`.

````python
from qiskit.quantum_info import Statevector

qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.cx(0, 1)
sv = Statevector(qc2)
print('Statevector:', sv.data)
print('Probabilities:', sv.probabilities_dict())
````

# 6. Endianness note

## Qiskit uses **little-endian** ordering: the qubit labelled `q0` is the **rightmost** bit in the output bitstring. So an outcome `'10'` means $q_1 = 1, q_0 = 0$. This catches everyone the first time — keep it in mind when interpreting counts.
