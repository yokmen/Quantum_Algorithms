# Quantum Walks

## Quantum walks are the quantum analogue of classical random walks. They spread quadratically faster than their classical counterparts and are the algorithmic engine behind element distinctness, triangle finding, spatial search, and other graph algorithms.

# 1. Discrete-time quantum walk

## On a graph with vertices $V$ and an additional **coin** Hilbert space, the walker's state lives in $\mathcal{H}_V \otimes \mathcal{H}_C$. One step:

## 1. Apply a unitary **coin operator** $C$ on the coin register (e.g. Hadamard for a 1D walk on $\mathbb Z$).
## 2. Apply the **shift operator** $S$ that moves the walker according to the coin.

## Repeat for $t$ steps. After $t$ steps, the walker on a line has spread to distance $\sim t$ (compared to $\sim \sqrt t$ classically).

```python
# 1D Hadamard walk on a line, simulated as a state vector.
import numpy as np

T = 30
N = 2*T + 1                                # positions -T..T
# Walker positions and coin (2-dim: left / right)
psi = np.zeros((N, 2), dtype=complex)
psi[T, 0] = 1                              # start at origin with coin in |0>

H = (1/np.sqrt(2))*np.array([[1, 1],[1, -1]])

for _ in range(T):
    psi = psi @ H.T                        # coin flip
    new = np.zeros_like(psi)
    new[1:, 0] = psi[:-1, 0]               # shift coin |0> right
    new[:-1, 1] = psi[1:, 1]               # shift coin |1> left
    psi = new

prob = np.sum(np.abs(psi)**2, axis=1)
print('Distribution after', T, 'steps (positions -T..T):')
print(prob.round(3))
```

# 2. Continuous-time quantum walk

## Drop the coin: evolve directly under $e^{-iAt}$ where $A$ is the (weighted) adjacency matrix of the graph. The walker's state lives in $\mathcal{H}_V$ alone — much closer to a Schrödinger evolution.

# 3. Quantum walk algorithms

## - **Spatial search**: find a marked vertex in a graph in $O(\sqrt N)$ time (Childs–Goldstone) — quadratic over classical   random walk hitting times.
## - **Element distinctness** (Ambainis): $O(N^{2/3})$ — strictly faster than Grover's $O(\sqrt N)$.
## - **Triangle finding** (Magniez–Santha–Szegedy): $O(N^{1.297})$ — also based on quantum walks on graphs of subsets.
## - **Quantum walk-based simulation** of physical systems on graphs.

# 4. Connection to amplitude amplification

## Many quantum-walk speedups can be understood as **walk-based amplitude amplification**: instead of resampling a subroutine, you take a step of a quantum walk that gradually pushes amplitude toward marked vertices. The Szegedy walk framework formalises this in a clean way.

# Recap

## - Two flavours: discrete-time (coin + shift) and continuous-time ($e^{-iAt}$).
## - Quadratic spreading vs classical random walks.
## - Algorithmic backbone for element distinctness, triangle finding, spatial search.
