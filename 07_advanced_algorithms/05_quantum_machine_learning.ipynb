# Quantum Machine Learning

## Quantum Machine Learning (QML) is an umbrella term covering several quite different ideas: quantum-enhanced versions of classical ML algorithms, classical algorithms inspired by quantum algorithms, and ML methods applied to quantum data. It is an extremely active and unsettled research area — what works and what doesn't is still being figured out.

# 1. The four corners

| Data       | Algorithm  | Examples                                |
|------------|------------|-----------------------------------------|
| Classical  | Classical  | ordinary ML                             |
| Classical  | Quantum    | Variational classifiers, QSVM           |
| Quantum    | Classical  | Tomography, error mitigation            |
| Quantum    | Quantum    | Quantum autoencoders, learning to simulate |

## Most "QML" today lives in the second row (classical data, quantum algorithm) using variational circuits.

# 2. Variational quantum classifiers

## Encode a classical input $\vec x$ into a quantum state $|\phi(\vec x)\rangle$, apply a parameterised circuit $U(\vec\theta)$, measure an observable $O$, and use the expectation value as the classifier output. The structure is identical to a (very small) neural network, with the parameterised quantum circuit playing the role of the network.

# 3. Quantum kernels

## An alternative angle: use quantum circuits to define a kernel

$$ \Large  k(\vec x, \vec y) = |\langle \phi(\vec x) | \phi(\vec y) \rangle|^2 $$

## and plug it into a classical SVM. The quantum part is just the kernel evaluation. For carefully constructed feature maps (Havlíček et al., 2019), one can prove that the kernel is hard to estimate classically — possibly opening a provable quantum learning advantage.

# 4. Quantum data

## When the data is *itself* quantum (e.g. states output by a physical system), quantum algorithms have a more obvious edge — they can process the data directly without expensive tomography. Examples:

## - **Quantum autoencoders**: compress a family of quantum states using a learned circuit.
## - **Quantum metrology**: learn an unknown unitary in fewer queries.
## - **Learning Hamiltonians**: infer interactions from observed dynamics.

# 5. Honest caveats

## - Many flashy early QML results turned out to be **dequantisable** (classical randomised algorithms with similar runtime).
## - Hardware-limited near-term QML demos are typically tiny — comparison to GPU-scale classical ML is generally unfavourable.
## - Barren plateaus (same as in VQE) plague variational QML.
## - A clear, end-to-end QML advantage on practical classical data remains elusive.

## That said, QML is one of the most active research areas in quantum computing and worth watching closely.

# Recap

## - QML spans four data × algorithm combinations; most current work uses classical data with variational quantum circuits.
## - Quantum kernels offer a clean route to provable separations on synthetic data.
## - For native quantum data (states from physical systems), the case for QML is much stronger.
