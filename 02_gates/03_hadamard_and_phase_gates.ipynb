# Hadamard, S and T Gates

## These three gates take us beyond "flip bits and signs". Hadamard creates superposition; $S$ and $T$ add phase. Together with CNOT (chapter 5), they form a universal gate set for quantum computation.

# 1. The Hadamard gate

$$ \Large  H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}. $$

## Action on basis states:

$$ \Large  H|0\rangle = |+\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \qquad H|1\rangle = |-\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle - |1\rangle). $$

## $H$ is its own inverse: $H^2 = I$. It is **the** gate that turns a definite bit into a balanced superposition, which is the starting move for nearly every quantum algorithm.

```python
import numpy as np
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)

ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)

print('H|0> =', H @ ket0)        # |+>
print('H|1> =', H @ ket1)        # |->
print('H^2 = I ?', np.allclose(H @ H, np.eye(2)))
```

# 2. The phase gate $S$

$$ \Large  S = \begin{pmatrix} 1 & 0 \\ 0 & i \end{pmatrix} = \sqrt{Z}. $$

## It leaves $|0\rangle$ alone and multiplies $|1\rangle$ by $i$. Two applications give $Z$ (a phase flip), so $S^2 = Z$ and $S^4 = I$. $S$ is what you reach for when you need an *imaginary* relative phase.

```python
S = np.array([[1, 0], [0, 1j]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)

print('S|0> =', S @ ket0)
print('S|1> =', S @ ket1)
print('S^2 = Z ?', np.allclose(S @ S, Z))
```

# 3. The $T$ gate

$$ \Large  T = \begin{pmatrix} 1 & 0 \\ 0 & e^{i\pi/4} \end{pmatrix} = \sqrt{S}. $$

## An eighth-root of the identity: $T^8 = I$. The $T$ gate is the smallest "non-Clifford" gate (see notebook 6 on universality). Adding it to the Clifford group is what lets you reach arbitrary single-qubit unitaries — and what makes fault-tolerant quantum computation hard, because $T$ gates are the bottleneck of fault-tolerant constructions.

```python
T = np.array([[1, 0], [0, np.exp(1j*np.pi/4)]], dtype=complex)

print('T^2 = S ?', np.allclose(T @ T, S))
print('T^8 = I ?', np.allclose(np.linalg.matrix_power(T, 8), np.eye(2)))
```

# 4. Sandwich identities

## $H$ swaps the $X$ and $Z$ Paulis:

$$ \Large  H X H = Z, \qquad H Z H = X, \qquad H Y H = -Y. $$

## In words: applying Hadamard before and after a $Z$ gate turns it into an $X$ gate. This is one of the most useful identities in quantum circuit design — it tells us we can implement a measurement in the Hadamard basis by applying $H$ first and then measuring in the computational basis.

```python
X = np.array([[0, 1], [1, 0]], dtype=complex)
Y = np.array([[0, -1j], [1j, 0]], dtype=complex)

print('HXH = Z ?', np.allclose(H @ X @ H, Z))
print('HZH = X ?', np.allclose(H @ Z @ H, X))
print('HYH = -Y ?', np.allclose(H @ Y @ H, -Y))
```

# 5. Why these three together

## The set $\{H, S\}$ together with CNOT (chapter 2 notebook 5) generates the **Clifford group** — efficiently simulable classically (Gottesman–Knill theorem). Adding $T$ extends this to a universal gate set:

$$ \Large  \{H, S, T, \text{CNOT}\} \;\Rightarrow\; \text{any unitary, to arbitrary accuracy}. $$

## We will see precisely what "universal" means in notebook 6.
