# Rotation Gates ($R_x, R_y, R_z$)

## Whereas $X, Y, Z, H, S, T$ are *discrete* gates, the rotation gates are **continuous families** parameterized by an angle. They describe arbitrary rotations of the Bloch vector around the three coordinate axes.

# 1. Definitions

## For any Pauli $P \in \{X, Y, Z\}$ and angle $\theta \in \mathbb{R}$,

$$ \Large  R_P(\theta) = e^{-i\theta P / 2} = \cos(\theta/2)\,I - i\sin(\theta/2)\,P. $$

## Explicitly:

$$ \Large  R_x(\theta) = \begin{pmatrix} \cos(\theta/2) & -i\sin(\theta/2) \\ -i\sin(\theta/2) & \cos(\theta/2) \end{pmatrix}, $$
$$ \Large  R_y(\theta) = \begin{pmatrix} \cos(\theta/2) & -\sin(\theta/2) \\ \sin(\theta/2) & \cos(\theta/2) \end{pmatrix}, $$
$$ \Large  R_z(\theta) = \begin{pmatrix} e^{-i\theta/2} & 0 \\ 0 & e^{i\theta/2} \end{pmatrix}. $$

```python
import numpy as np

def Rx(theta):
    c, s = np.cos(theta/2), np.sin(theta/2)
    return np.array([[c, -1j*s], [-1j*s, c]])

def Ry(theta):
    c, s = np.cos(theta/2), np.sin(theta/2)
    return np.array([[c, -s], [s, c]])

def Rz(theta):
    return np.array([[np.exp(-1j*theta/2), 0], [0, np.exp(1j*theta/2)]])

# Sanity check: Rx(0) = I
print('Rx(0) =\n', Rx(0).round(10))
```

# 2. The $\theta/2$ convention

## Why does a rotation by angle $\theta$ contain $\theta/2$ in its matrix? Because a qubit is a **spin-$1/2$ system**: rotating it by $2\pi$ takes the state vector to $-|\psi\rangle$ (which is the same physical state, but with a global phase of $-1$). A full $4\pi$ rotation returns it to itself exactly. This is the famous *spinor* behaviour.

```python
# Demonstrate the spinor 4*pi periodicity
print('Rx(2*pi) =\n', Rx(2*np.pi).round(10))     # = -I
print('Rx(4*pi) =\n', Rx(4*np.pi).round(10))     # = +I
```

# 3. Relation to the Pauli gates

## Pauli gates are rotations by $\pi$, up to a global phase:

$$ \Large  R_x(\pi) = -i X, \quad R_y(\pi) = -i Y, \quad R_z(\pi) = -i Z. $$

## Similarly $S = e^{i\pi/4}\,R_z(\pi/2)$ and $T = e^{i\pi/8}\,R_z(\pi/4)$. Up to global phase, $S$ and $T$ are $z$-rotations by $\pi/2$ and $\pi/4$ respectively.

```python
X = np.array([[0, 1], [1, 0]], dtype=complex)
print('Rx(pi) =\n', Rx(np.pi).round(10))
print('-iX   =\n', (-1j * X).round(10))
print('match? ', np.allclose(Rx(np.pi), -1j * X))
```

# 4. Decomposing arbitrary single-qubit gates

## **ZYZ decomposition**: any single-qubit unitary $U$ can be written (up to global phase) as

$$ \Large  U = R_z(\alpha)\, R_y(\beta)\, R_z(\gamma) $$

## for some real $\alpha, \beta, \gamma$. This means rotations around just two axes ($y$ and $z$) are enough to build any single-qubit operation. Compilers for real hardware use exactly this fact to reduce gate sets.

```python
# Verify ZYZ decomposition for the Hadamard
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)
decomp = Rz(0) @ Ry(np.pi/2) @ Rz(np.pi)
# Account for the global phase factor between the two
phase = H[0,0] / decomp[0,0]
print('H matches Rz(0) Ry(pi/2) Rz(pi) up to global phase?',
      np.allclose(H, phase * decomp))
```

# 5. Why rotations matter for algorithms

## - **Variational algorithms** (VQE, QAOA — chapter 07) use rotations as their *tunable parameters*.   The angles are optimised to minimise an energy or cost function.
## - **Quantum signal processing** and **block-encoding** are built on long products of rotation gates.
## - **Hardware-native gates** on superconducting qubits are typically $R_z$ (virtual, free) plus one fixed entangling gate.
