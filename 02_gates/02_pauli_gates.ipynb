# Pauli Gates (X, Y, Z)

## The three **Pauli matrices** are arguably the most important operators in all of quantum mechanics. They serve simultaneously as gates (unitaries) and as observables (Hermitian operators).

# 1. The matrices

$$ \Large  X = \begin{pmatrix}0 & 1\\ 1 & 0\end{pmatrix},\quad Y = \begin{pmatrix}0 & -i\\ i & 0\end{pmatrix},\quad Z = \begin{pmatrix}1 & 0\\ 0 & -1\end{pmatrix}. $$

## Each is **Hermitian** ($P^\dagger = P$) and **unitary** ($P^\dagger P = I$), and each squares to the identity ($P^2 = I$). Their eigenvalues are therefore $\pm 1$.

```python
import numpy as np
X = np.array([[0,1],[1,0]], dtype=complex)
Y = np.array([[0,-1j],[1j,0]], dtype=complex)
Z = np.array([[1,0],[0,-1]], dtype=complex)

for name, P in [('X',X),('Y',Y),('Z',Z)]:
    print(f'{name}^2 = I ? {np.allclose(P @ P, np.eye(2))},  Hermitian? {np.allclose(P, P.conj().T)}')
```

# 2. Action on the computational basis

## - $X$ is the **quantum NOT**: $X|0\rangle = |1\rangle$, $X|1\rangle = |0\rangle$.
## - $Z$ is the **phase flip**: $Z|0\rangle = |0\rangle$, $Z|1\rangle = -|1\rangle$.
## - $Y = iXZ$ combines both: $Y|0\rangle = i|1\rangle$, $Y|1\rangle = -i|0\rangle$.

```python
ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)

print('X|0> =', X @ ket0, '   X|1> =', X @ ket1)
print('Z|0> =', Z @ ket0, '   Z|1> =', Z @ ket1)
print('Y|0> =', Y @ ket0, '   Y|1> =', Y @ ket1)
```

# 3. The algebra of the Paulis

## The Paulis **anticommute** with each other:

$$ \Large  \{X, Y\} = XY + YX = 0, \quad \{Y, Z\} = 0, \quad \{X, Z\} = 0. $$

## And they multiply cyclically:

$$ \Large  XY = iZ, \quad YZ = iX, \quad ZX = iY. $$

## These relations, plus $X^2 = Y^2 = Z^2 = I$, are the **Pauli algebra** — it shows up everywhere in quantum information.

```python
# Verify Pauli algebra
print('XY = iZ ?', np.allclose(X @ Y, 1j*Z))
print('YZ = iX ?', np.allclose(Y @ Z, 1j*X))
print('ZX = iY ?', np.allclose(Z @ X, 1j*Y))
print('XY + YX = 0 ?', np.allclose(X @ Y + Y @ X, 0))
```

# 4. Why these three?

## Together with the identity, $\{I, X, Y, Z\}$ form an **orthogonal basis** for the real vector space of $2 \times 2$ Hermitian matrices (under the Hilbert–Schmidt inner product $\langle A, B\rangle = \tfrac{1}{2}\mathrm{Tr}(A^\dagger B)$). So any single-qubit observable can be expanded as

$$ \Large  A = a_0 I + a_x X + a_y Y + a_z Z, \qquad a_i \in \mathbb{R}. $$

## On the Bloch sphere (notebook 7), the three Paulis correspond exactly to measurements along the $x, y, z$ axes.

```python
# Expand the Hadamard in the Pauli basis: H = (X + Z)/sqrt(2)
H = (1/np.sqrt(2)) * np.array([[1,1],[1,-1]], dtype=complex)
expansion = (X + Z) / np.sqrt(2)
print('H = (X + Z)/sqrt(2) ?', np.allclose(H, expansion))
```

# 5. Pauli observables and measurement

## Measuring the **Z observable** is the same thing as measuring in the computational basis: outcome $+1$ collapses to $|0\rangle$, outcome $-1$ collapses to $|1\rangle$.

## Measuring **X** is measuring in the $\{|+\rangle, |-\rangle\}$ basis; measuring **Y** is measuring in the $\{|{+i}\rangle, |{-i}\rangle\}$ basis where $|{\pm i}\rangle = (|0\rangle \pm i|1\rangle)/\sqrt{2}$. All three together completely characterise a single qubit's state — they are how a **state tomography** experiment works.
