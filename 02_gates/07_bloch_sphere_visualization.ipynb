# Bloch Sphere Visualization

## Every (pure) qubit state, modulo global phase, corresponds to a point on the surface of a unit sphere — the **Bloch sphere**. Single-qubit gates are then rotations of that sphere. This picture is the single most useful tool for building intuition about qubit dynamics.

# 1. Coordinates

## Write a qubit state (up to global phase) as

$$ \Large  |\psi\rangle = \cos(\theta/2)|0\rangle + e^{i\varphi}\sin(\theta/2)|1\rangle. $$

## The pair $(\theta, \varphi)$ are the polar and azimuthal angles of a point on the unit sphere:

$$ \Large  \vec r = (\sin\theta\cos\varphi,\; \sin\theta\sin\varphi,\; \cos\theta). $$

## The six "compass points" of the sphere correspond to the six common basis states:

## - $+z$: $|0\rangle$, $\;\;-z$: $|1\rangle$
## - $+x$: $|+\rangle$, $\;-x$: $|-\rangle$
## - $+y$: $|{+}i\rangle$, $-y$: $|{-}i\rangle$

```python
import numpy as np

def bloch_vector(psi):
    psi = psi / np.linalg.norm(psi)
    X = np.array([[0,1],[1,0]], dtype=complex)
    Y = np.array([[0,-1j],[1j,0]], dtype=complex)
    Z = np.array([[1,0],[0,-1]], dtype=complex)
    rx = (psi.conj() @ X @ psi).real
    ry = (psi.conj() @ Y @ psi).real
    rz = (psi.conj() @ Z @ psi).real
    return np.array([rx, ry, rz])

for label, psi in [
    ('|0>',   np.array([1, 0])),
    ('|1>',   np.array([0, 1])),
    ('|+>',   np.array([1, 1])/np.sqrt(2)),
    ('|->',   np.array([1, -1])/np.sqrt(2)),
    ('|+i>',  np.array([1, 1j])/np.sqrt(2)),
    ('|-i>',  np.array([1, -1j])/np.sqrt(2)),
]:
    print(f'{label:5s}  ->  {bloch_vector(psi).round(3)}')
```

# 2. Gates as rotations

## Single-qubit gates correspond to rotations of the Bloch sphere:

## - $R_x(\theta)$ rotates by $\theta$ around the $x$-axis.
## - $R_y(\theta)$ rotates by $\theta$ around the $y$-axis.
## - $R_z(\theta)$ rotates by $\theta$ around the $z$-axis.
## - $X, Y, Z$ are rotations by $\pi$ around the respective axes.
## - $H$ is a rotation by $\pi$ around the axis $(\hat x + \hat z)/\sqrt 2$ — it swaps the $x$ and $z$ axes.

```python
# Demonstrate: H takes |0> (north pole) to |+> (positive x axis)
H = (1/np.sqrt(2))*np.array([[1,1],[1,-1]], dtype=complex)
psi = np.array([1, 0], dtype=complex)
psi_after = H @ psi
print('Bloch vector of |0>      :', bloch_vector(psi).round(3))
print('Bloch vector of H|0>     :', bloch_vector(psi_after).round(3))
```

# 3. A 2D rendering of the sphere

## Matplotlib renders the sphere with the state vector. Changing the gate or state in the code below helps build intuition.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # noqa: F401

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# wireframe sphere
u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, np.pi, 20)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones_like(u), np.cos(v))
ax.plot_wireframe(x, y, z, color='gray', alpha=0.2)

# axes
for vec, lab in [((1.2,0,0),'x'),((0,1.2,0),'y'),((0,0,1.2),'z')]:
    ax.plot([0, vec[0]],[0, vec[1]],[0, vec[2]], 'k')
    ax.text(*vec, lab)

# the state vector for |+>
r = bloch_vector(np.array([1, 1])/np.sqrt(2))
ax.quiver(0, 0, 0, *r, color='red', linewidth=2)
ax.set_box_aspect([1, 1, 1])
ax.set_title('Bloch sphere: |+>')
plt.show()
```

# 4. Mixed states and the interior

## We have only described **pure** states, which live on the *surface* of the sphere. A **mixed state** (a probability distribution over pure states) corresponds to a point in the *interior*. The center is the maximally mixed state $I/2$ — the qubit equivalent of a fair coin.

## Distance from the center is $|\vec r| \in [0, 1]$: $1$ for pure, $0$ for maximally mixed. We will see more of mixed states when we discuss noise in chapter 9.

# Recap

## - Pure qubit states live on the surface of a unit sphere; mixed states fill the interior.
## - $|0\rangle, |1\rangle$ are the poles; $|\pm\rangle, |\pm i\rangle$ are the equatorial points.
## - Gates rotate the Bloch sphere; rotation gates are exactly $R_P$.
## - The Bloch picture is the most useful intuition for single-qubit dynamics — worth keeping in mind constantly.
