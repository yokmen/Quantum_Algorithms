# Single-Qubit Gates

## A **quantum gate** is a unitary matrix that acts on the state vector of a qubit (or several qubits). Single-qubit gates are $2 \times 2$ unitaries acting on $\mathbb{C}^2$. They are the elementary operations from which every quantum circuit is built.

# 1. Why unitary?

## Recall from chapter 1 that quantum states are unit vectors. A unitary $U$ satisfies $U^\dagger U = I$, which means it preserves the norm:

$$ \Large  \|U|\psi\rangle\| = \sqrt{\langle\psi|U^\dagger U|\psi\rangle} = \sqrt{\langle\psi|\psi\rangle} = 1. $$

## So unitaries map valid quantum states to valid quantum states. They are also invertible (since $U^{-1} = U^\dagger$) — quantum evolution is **reversible**, in stark contrast to classical irreversible gates like AND or OR.

# 2. The general single-qubit unitary

## Every $2 \times 2$ unitary can be written (up to global phase) as

$$ \Large  U = \begin{pmatrix} \cos(\theta/2) & -e^{i\lambda}\sin(\theta/2) \\ e^{i\varphi}\sin(\theta/2) & e^{i(\varphi+\lambda)}\cos(\theta/2) \end{pmatrix}. $$

## Three real parameters $(\theta, \varphi, \lambda)$ are enough to describe any single-qubit gate. This is the **U3 gate** in Qiskit's nomenclature.

```python
import numpy as np

def U3(theta, phi, lam):
    c, s = np.cos(theta/2), np.sin(theta/2)
    return np.array([[c,                       -np.exp(1j*lam)*s],
                     [np.exp(1j*phi)*s,         np.exp(1j*(phi+lam))*c]])

U = U3(np.pi/3, np.pi/4, np.pi/5)
print('unitary?', np.allclose(U.conj().T @ U, np.eye(2)))
```

# 3. The standard zoo of single-qubit gates

## Most calculations only use a handful of named gates:

## - $I$ — identity (do nothing).
## - $X$, $Y$, $Z$ — the **Pauli gates** (notebook 2).
## - $H$ — the **Hadamard** (notebook 3).
## - $S$, $T$ — phase gates of orders 4 and 8 (notebook 3).
## - $R_x(\theta)$, $R_y(\theta)$, $R_z(\theta)$ — **rotations** around Bloch axes (notebook 4).

## We will spend the next four notebooks getting fluent with these.

```python
# Define the most common single-qubit gates as numpy matrices
I = np.eye(2, dtype=complex)
X = np.array([[0, 1], [1, 0]], dtype=complex)
Y = np.array([[0, -1j], [1j, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)
S = np.array([[1, 0], [0, 1j]], dtype=complex)
T = np.array([[1, 0], [0, np.exp(1j*np.pi/4)]], dtype=complex)

for name, G in [('I',I),('X',X),('Y',Y),('Z',Z),('H',H),('S',S),('T',T)]:
    print(f'{name} unitary? {np.allclose(G.conj().T @ G, np.eye(2))}')
```

# 4. How a gate acts on a state

## Applying gate $U$ to state $|\psi\rangle$ just means matrix multiplication: $|\psi'\rangle = U|\psi\rangle$. If the state is in superposition, all amplitudes are transformed simultaneously — this is the linearity that quantum algorithms exploit.

```python
ket0 = np.array([1, 0], dtype=complex)
ket1 = np.array([0, 1], dtype=complex)

print('X|0> =', X @ ket0)            # = |1>
print('H|0> =', H @ ket0)            # = |+>
print('Z|+> =', Z @ (H @ ket0))      # = |->
```

# 5. Composing gates

## If you apply $U_1$ then $U_2$, the combined operation is the matrix product $U_2 U_1$ — **right-to-left**, because gates act on kets to their right. This is the opposite of how circuit diagrams are read.

$$ \Large  \text{circuit: } |\psi\rangle \;\to\; U_1 \;\to\; U_2 \;\to\; U_2 U_1 |\psi\rangle $$

```python
# H then Z then H should equal X (a famous identity)
HZH = H @ Z @ H
print('H Z H =\n', np.round(HZH, 10))
print('equals X?', np.allclose(HZH, X))
```
