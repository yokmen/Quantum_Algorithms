# The Stabilizer Formalism

## Almost every useful QEC code is a **stabilizer code**: defined as the simultaneous +1 eigenspace of a commuting group of Pauli operators. This algebraic framing turns QEC from a sea of ad-hoc constructions into a clean, computable theory.

# 1. Stabilizers

## Let $\mathcal{P}_n$ be the $n$-qubit Pauli group. A **stabilizer group** $S \subseteq \mathcal{P}_n$ is an abelian subgroup not containing $-I$. The **code space** is

$$ \Large  \mathcal C = \{ |\psi\rangle : g|\psi\rangle = +|\psi\rangle \text{ for all } g \in S\}. $$

## If $S$ is generated by $n - k$ independent stabilizers, $\dim \mathcal C = 2^k$, so the code encodes $k$ logical qubits.

# 2. Example: the 3-qubit code

## The bit-flip code from notebook 2 has stabilizer generators $\{ Z_0 Z_1, Z_1 Z_2 \}$. The code space is

$$ \Large  \{\alpha|000\rangle + \beta|111\rangle\}. $$

## Both states have eigenvalue $+1$ under $Z_0 Z_1$ and $Z_1 Z_2$ — they are exactly the $+1$ eigenspace.

## Logical operators are Paulis that commute with all stabilizers but are not themselves in the stabilizer group. For the bit-flip code, $\bar X = X_0 X_1 X_2$ and $\bar Z = Z_0$ (or $Z_1$, or $Z_2$ — any one works).

```python
# Verify the stabilizers of the 3-qubit code
import numpy as np

X = np.array([[0,1],[1,0]], dtype=complex); Z = np.array([[1,0],[0,-1]], dtype=complex); I = np.eye(2, dtype=complex)
def tensor(ops):
    out = ops[0]
    for o in ops[1:]:
        out = np.kron(out, o)
    return out

ZZI = tensor([Z, Z, I])
IZZ = tensor([I, Z, Z])

for alpha, beta in [(1, 0), (0, 1), (1/np.sqrt(2), 1/np.sqrt(2))]:
    psi = alpha*tensor([np.array([1,0]), np.array([1,0]), np.array([1,0])]) + \
          beta *tensor([np.array([0,1]), np.array([0,1]), np.array([0,1])])
    print(f'alpha={alpha:.2f}, beta={beta:.2f}: ZZI psi - psi = {np.linalg.norm(ZZI @ psi - psi):.2e}')
    print(f'                            : IZZ psi - psi = {np.linalg.norm(IZZ @ psi - psi):.2e}')
```

# 3. Why it's powerful

## - **Compact description**: a stabilizer code on $n$ qubits is specified by $n - k$ Paulis, total $O(n)$ bits — vs.   $2^n$ complex amplitudes for an arbitrary state.
## - **Efficient simulation**: stabilizer states can be tracked classically in $O(n^2)$ time using the **tableau algorithm**   (Gottesman–Knill theorem).
## - **Error detection** becomes simple: measure each stabilizer generator; the bit pattern of $\pm 1$ outcomes is the **syndrome**.
## - The framework cleanly handles single-qubit Clifford gates: they map Paulis to Paulis, so a stabilizer code stays a stabilizer code under Clifford operations.

# 4. CSS codes

## A particularly clean subfamily: **CSS codes** (Calderbank–Shor–Steane) are stabilizer codes whose generators split into two sets — pure $X$-type and pure $Z$-type stabilizers — derived from two classical linear codes $C_X \supseteq C_Z^\perp$. The Steane code, surface codes, and Reed–Muller codes are all CSS.

## CSS codes inherit transversal $H$ and CNOT gates, making them especially nice for fault tolerance.
